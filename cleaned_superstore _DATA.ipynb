```python
import pandas as pd
```

```python
st = pd.read_csv("sampleSuperstore.csv")
st.head()
```

```python
st["Sales"].dtypes
```

```python
st.isnull().sum()
```

```python
st.notna()
```

```python
# st.dropna()
empty_row = st.dropna(subset = ["Sales"])
print(empty_row)
```

```python
st.head(15)
```

```python
print("total rows before cleaning", st.shape[0])
```

```python
st.duplicated().sum()
```

```python
st.drop_duplicates(inplace = True)
```

```python
st.duplicated().sum()
```

```python
print("total_rows after cleaning", st.shape[0])
```

```python
duplicate_rows = st[st.duplicated()]
print(duplicate_rows.head())
```

```python
Region_sales = st.groupby("Region")["Sales"].sum()
print("high sales in the",Region_sales)

Region_low_sales = st.groupby("Region")["Sales"].sum().sort_values().head(2)
print("low sales in the ",Region_low_sales)
```

```python
State_high_sales = st.groupby("State")["Sales"].sum().head(5).sort_values(ascending = True)
print("High sales in the", State_high_sales)

State_low_sales = st.groupby("State")["Sales"].sum().head(3).sort_values(ascending = True)
print("low sales in the", State_low_sales)
```

```python
category_sales = st.groupby("Category")["Sales"].sum().head().sort_values(ascending=False)
print(category_sales)
```

```python
top_10_product = st.groupby("Sub-Category")["Sales"].sum().head(5).sort_values(ascending = False)
print(top_10_product)
```

```python
correlation = st["Discount"].corr(st["Profit"])
print("Due to discount we come profit or loss", correlation)
```

```python
shipping_analysis = st.groupby(["Ship Mode", "Region"]).size().unstack()
print(shipping_analysis)
```

```python
st["Sales"] = st["Sales"].fillna(0)
print(st["Sales"])
```
